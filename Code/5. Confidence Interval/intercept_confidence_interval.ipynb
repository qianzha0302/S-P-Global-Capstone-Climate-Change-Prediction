```python
import pandas as pd
import numpy as np
import scipy.stats as st
```

###CA

```python
#https://drive.google.com/file/d/1HaJg7ZElravEm3Sirsd4abeG8_CFohAH/view?usp=share_link

id = "1HaJg7ZElravEm3Sirsd4abeG8_CFohAH"
file = drive.CreateFile({'id':id}) 
file.GetContentFile('ca intercept table.csv')
```

```python
ca_intercept = pd.read_csv('ca intercept table.csv')
```

```python
ca_intercept
```

```python
ca_126_winter = ca_intercept['126Winter INTCP']
ca_126_spring = ca_intercept['126Spring INTCP']
ca_126_fall = ca_intercept['126Fall INTCP']
ca_126_summer = ca_intercept['126Summer INTCP']
ca_585_winter = ca_intercept['585Winter INTCP']
ca_585_spring = ca_intercept['585Spring INTCP']
ca_585_fall = ca_intercept['585Fall INTCP']
ca_585_summer = ca_intercept['585Summer INTCP']
```

```python
ca_126_winter_interval = st.t.interval(alpha=0.95, df=len(ca_126_winter)-1, loc=np.mean(ca_126_winter), scale=st.sem(ca_126_winter)) 
ca_126_spring_interval = st.t.interval(alpha=0.95, df=len(ca_126_spring)-1, loc=np.mean(ca_126_spring), scale=st.sem(ca_126_spring)) 
ca_126_fall_interval = st.t.interval(alpha=0.95, df=len(ca_126_fall)-1, loc=np.mean(ca_126_fall), scale=st.sem(ca_126_fall)) 
ca_126_summer_interval = st.t.interval(alpha=0.95, df=len(ca_126_summer)-1, loc=np.mean(ca_126_summer), scale=st.sem(ca_126_summer)) 
```

```python
ca_585_winter_interval = st.t.interval(alpha=0.95, df=len(ca_585_winter)-1, loc=np.mean(ca_585_winter), scale=st.sem(ca_585_winter)) 
ca_585_spring_interval = st.t.interval(alpha=0.95, df=len(ca_585_spring)-1, loc=np.mean(ca_585_spring), scale=st.sem(ca_585_spring)) 
ca_585_fall_interval = st.t.interval(alpha=0.95, df=len(ca_585_fall)-1, loc=np.mean(ca_585_fall), scale=st.sem(ca_585_fall)) 
ca_585_summer_interval = st.t.interval(alpha=0.95, df=len(ca_585_summer)-1, loc=np.mean(ca_585_summer), scale=st.sem(ca_585_summer)) 
```

```python
print(ca_126_winter_interval) 
print(ca_126_spring_interval)  
print(ca_126_fall_interval)  
print(ca_126_summer_interval)
```

```python
print(ca_585_winter_interval) 
print(ca_585_spring_interval)  
print(ca_585_fall_interval)  
print(ca_585_summer_interval)
```
