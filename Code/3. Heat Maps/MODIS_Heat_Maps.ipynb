```python
data = Dataset(r"output.nc")
```

```python
data.variables.keys() 
```

```python
lats=data.variables['lat'][:]
lons=data.variables['lon'][:]
time=data.variables['time'][:]
crs=data.variables['crs'][:]
tmp=data.variables['LST_Day_1km'][:]
data.variables['LST_Day_1km']
```

```python
tmp_c=tmp-273.15 # K to C
```

```python
time
```

```python
mp = Basemap(projection='merc',
            llcrnrlon=-119.9,
            llcrnrlat=36.25770090268727,
            urcrnrlon=-117,
            urcrnrlat=39.11066973350914,
            resolution = 'i')
```

```python
lon,lat=np.meshgrid(lons, lats)
x,y=mp(lon,lat)
plt.figure(figsize=(10,10))
```

```python
days=np.arange(0,12)
for i in days: 
    c_scheme=mp.pcolor(x,y,np.squeeze(tmp_c[i,:,:]),cmap='jet')
    
    cbar = mp.colorbar(c_scheme,location='right',pad = '10%')
    
    meridians= np.arange(-120.14568539999999,-117.03655454062499)
    mp.drawmeridians(meridians,labels=[0,0,0,1])
    
    parallels = np.arange(36.25770090268727,39.27066973350914)
    mp.drawparallels(parallels,labels=[1,0,0,0])
    
    day=i+1
    
    if day==1:
        plt.title('CA Land-Surface Temperature for Jan')
                  
    elif day==2:
        plt.title('CA Land-Surface Temperature for Feb')
        
    elif day==3:
        plt.title('CA Land-Surface Temperature for Mar')
    
    elif day==4:
        plt.title('CA Land-Surface Temperature for Apr')
    
    elif day==5:
        plt.title('CA Land-Surface Temperature for May')
    
    elif day==6:
        plt.title('CA Land-Surface Temperature for Jun')
    
    elif day==7:
        plt.title('CA Land-Surface Temperature for Jul')
        
    elif day==8:
        plt.title('CA Land-Surface Temperature for Aug')
        
    elif day==9:
        plt.title('CA Land-Surface Temperature for Sep')
        
    elif day==10:
        plt.title('CA Land-Surface Temperature for Oct')
        
    elif day==11:
        plt.title('CA Land-Surface Temperature for Nov')
        
    elif day==12:
        plt.title('CA Land-Surface Temperature for Dec')

    plt.clim(-20,50)
    
    plt.savefig(r"MOD11A1.2015_"+str(day)+'.jpg')
    plt.clf()
```

```python
import PIL

image_frames = [] 
month = np.arange(1,12)

for k in month:
    new_fram = PIL.Image.open(r"MOD11A1.2015_" + str(k) + '.jpg') 
    image_frames.append(new_fram)

image_frames[0].save('MODIS.gif',format='GIF',
                    append_images = image_frames[1: ],
                    save_all = True, duration = 100,
                    loop = 0)
```
