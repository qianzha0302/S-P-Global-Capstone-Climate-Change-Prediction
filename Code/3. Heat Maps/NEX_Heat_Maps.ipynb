```python
from netCDF4 import Dataset
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import xarray
import pandas as pd
import os
```

###Turk_BCC-CSM2 & CESM2

```python
data = Dataset(r"MOD11A1.061_1km_aid0001.nc")
```

```python
data.variables.keys()
```

```python
lats=data.variables['lat'][:]
lons=data.variables['lon'][:]
time=data.variables['time'][:]
crs=data.variables['crs'][:]
tmp=data.variables['LST_Day_1km'][:]
data.variables['LST_Day_1km']
```

```python
tmp_c=tmp-273.15
```

```python
tmp_c
```

```python
time
```

```python
mp = Basemap(projection='merc',
            llcrnrlon=-120.00000,
            llcrnrlat=36.00000,
            urcrnrlon=-117.00000,
            urcrnrlat=39.00000,
            resolution = 'i')
```

```python
lon,lat=np.meshgrid(lons,lats)
x,y=mp(lon,lat)
plt.figure(figsize=(10,10))
```

```python
c_scheme=mp.pcolor(x,y,np.squeeze(tmp_c[0,:,:]),cmap='jet')

    
cbar = mp.colorbar(c_scheme,location='right',pad = '10%')
    
meridians= np.arange(-120.14568539999999,-117.03655454062499)
mp.drawmeridians(meridians,labels=[0,0,0,1])
    
parallels = np.arange(36.25770090268727,39.27066973350914)
mp.drawparallels(parallels,labels=[1,0,0,0])

plt.title('Turk_ssp585_CESM2 Near-Surface Air Temperature for Day 1')

plt.clim(-15,15)
    
plt.savefig(r"Heatmap"+'\\'+str(1)+'.jpg')
plt.clf()
```
